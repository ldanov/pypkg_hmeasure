```python
from scipy.stats import beta as beta_dist
from scipy.special import beta as beta_func
from scipy.special import betainc as betainc_func
import numpy
import matplotlib.pyplot as plt
```

```python
prange = numpy.arange(0.01, 0.999, 0.01)
```

```python
prange
```

```python
beta_inst = beta_dist(2, 2)
```

```python
beta_ppf = beta_inst.ppf(prange)
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(prange, beta_ppf)
```

```python
# scipy.stats.beta.cdf == pbeta
beta_cdf = beta_inst.cdf(prange)
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(prange, beta_cdf)
```

```python
# scipy.stats.beta.pdf == dbeta
beta_pdf = beta_inst.pdf(prange)
fig, ax = plt.subplots(1, 1)
ax.plot(prange, beta_pdf)
```
